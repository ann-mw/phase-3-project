# data preparation

## Data cleaning

```python
#confirming that there no missing values (nan) in the dataframe
missing_values = df.isnull().sum()
print(missing_values)
```

```python
print("Original data shape:", df.shape)
print("\nNumber of features:", len(df.columns))
print("\nColumns:")
print(df.columns)
```

```python
# Separate numerical and categorical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
# Create transformers
numerical_transformer = SimpleImputer(strategy='mean')
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, num_cols),
        ('cat', categorical_transformer, cat_cols)
    ]
)
```

```python
# Fit and transform the data
X_prepared = preprocessor.fit_transform(df.drop(['churn'], axis=1))
```

```python
# Manually create feature names
feature_names = []
for col in df.columns[:-1]:
    if col in num_cols:
        feature_names.append(f"{col}_mean")
    else:
        # Count unique values in each categorical column
        unique_values = df[col].unique()
        feature_names.extend([f"{col}_{value}" for value in unique_values])
```

```python
# Create a new DataFrame with the prepared data and feature names
df_prepared = pd.DataFrame(X_prepared, columns=feature_names)
```

```python
# Handle the 'churn' column
if 'churn' in df.columns:
    # Convert boolean values to integers (True -> 1, False -> 0)
    df_prepared['churn'] = df['churn'].astype(int)
```
